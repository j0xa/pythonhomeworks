# Task 1

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(-10, 10, 10000)
y1 = x1**2-4*x1+4

plt.plot(x1, y1, label=r"$f(x) = x^2 - 4x + 4$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc="upper right")
plt.show()
```

# Task 2

```python
x2 = np.linspace(0, np.pi, 1000)
y2_1 = np.sin(x2)
y2_2 = np.cos(x2)

plt.plot(x2, y2_1, color="red", linestyle=":", label=r"$ \sin(x) $")
plt.plot(x2, y2_2, color="blue", linestyle="--", label=r"$ \cos(x) $")
plt.legend()
plt.show()
```

# Task 3

```python
x3 = np.linspace(0, 10, 1000)
top_left = x3**3
top_right = np.sin(x3)
bottom_left = np.e**x3
bottom_right = np.where(x3!=0, np.log(x3+1),0)
plt.subplot(2, 2, 1)
plt.plot(x3, top_left, color="black", label=r"$ f(x) = x^3 $")
plt.legend()
plt.subplot(2, 2, 2)
plt.plot(x3, top_right, color="red", label=r"$ f(x) = \sin(x) $")
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(x3, bottom_left, color="blue", label=r"$ f(x) = e^x $")
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(x3, bottom_right, color="green", label=r"$ f(x) = \log(x+1) $ (for $ x \geq 0 $)")
plt.legend()
plt.show()
```

# Task 4

```python
# make a list of random numbers
x4 = np.random.uniform(0, 10, 100)
y4 = np.random.uniform(0, 10, 100)

# Make a list of random markers
markers = ['o', 's', 'D', 'v', '^', '<', '>', 'p', '*', 'h']
marker_choices = np.random.choice(markers, size=100)
plt.figure(figsize=(8, 6))

# Scatter the points
for i in range(100):
    plt.scatter(x4[i], y4[i], marker=marker_choices[i])

# Add Title and axis labels
plt.title("Scatter Plot of 100 Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Add grid
plt.grid(True, linestyle="--", alpha=0.7)
# Show the plot
plt.show()
```

# Task 5

```python
values = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(values, bins=30, color="red", edgecolor="black", alpha=0.7)
plt.title("Task 5")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

# Task 6

```python
x6 = np.linspace(-5, 5, 1000)
y6 = np.linspace(-5, 5, 1000)


xx, yy = np.meshgrid(x6, y6)
f6 = np.cos(xx**2+yy**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(xx, yy, f6, cmap="viridis")
plt.title(r"$ f(x, y) = \cos(x^2 + y^2) $")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Task 7

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = [200, 150, 250, 175, 225]
bar_colors = ["red", "orange", "yellow", "green", "blue"]

plt.bar(products, values, color=bar_colors, width=0.6, edgecolor="black")
plt.ylabel("Sales Values")
plt.xlabel("Products")
plt.title("Sales Data")
plt.show()
```

```python
values_a = [4, 2, 1, 3]
values_b = [7, 3, 5, 8]
values_c = [3, 6, 4, 5]
time_periods = ["T1", "T2", "T3", "T4"]

plt.bar(time_periods, values_a, label="Category A", color="red")
plt.bar(time_periods, values_b, bottom=values_a, label="Category B", color="green")
plt.bar(time_periods, values_c, bottom=np.array(values_a)+np.array(values_b), label="Category C", color="blue")
plt.legend()
plt.xlabel("Time Periods")
plt.ylabel("Values")
plt.title("Stacked Bar Chart of Categories over Time")
plt.show()
```
